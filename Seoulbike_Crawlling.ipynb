# 따릉이 홈페이지 크롤링

## 고객의소리에서 불만사항 파악하기

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

### 따릉이 고객의소리페이지 html로 가져오기

```python
html = urlopen("https://www.bikeseoul.com/customer/opinionBoard/opinionBoardList.do")  
bsObject = BeautifulSoup(html, "html.parser") 
bsObject
```

### 따릉이홈페이지 title 확인

```python
bsObject.head.title
```

### Html의 body 속성이 td 안의 class = "left" 찾기

```python
print(bsObject.body.find("td",{"class":"left"}))
```

#### 해당텍스트 찾기

```python
review = bsObject.find('td',{'class':'left'})
review.get_text().strip()
```

```python
for link in bsObject.findAll('td',{'class':'left'}):
    print(link.get_text().strip())
```

```python
base_url = 'https://www.bikeseoul.com/customer/opinionBoard/opinionBoardList.do?currentPageNo={}'

for n in range(1739):
    url = base_url.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html.parser')
    reviews = source.find_all('td',{'class':'left'})
    
    for review in reviews :
        print(review.get_text().strip())
```
